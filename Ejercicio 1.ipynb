```python
import random #líbrería para generar número aleatorios en [0,1]

def volados():   #hacemos la función que hará la simulación de los 1000 volados
    lvolados=[]   #hacemos una lista donde se guardarán los resultados de la simulación
    n=0  #contador
    sol=0
    aguila=0
    
    while n < 1000:    # bucle que hará las 1000 repeticiones
        x=random.random()  #generamos el número aleatorio 
        if x < 0.5:    # condición si x es menor que 0.5 el resultado será sol
            lvolados.append("Sol")  #escribimos en la lista el resultado
            sol += 1
        else:
            lvolados.append("Aguila") #en caso contrario, el resultado es águila y se agrega a la lista
            aguila += 1
        n += 1  #contador aumentando en una unidad
    P_sol=sol/n
    return lvolados, len(lvolados), P_sol   #regresa la lista, el total de resultados simulados, la prob de sol
```
