##  INTEGRAL DE LA RAÍZ DE ARCTAN(x)

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import uniform as unif

cant_num = 10000    #Cantidad de números aleatorios a generar

c = np.linspace(0,1, cant_num)
f = np.sqrt(np.arctan(c)) # definimos la función

lim_sup = 1   #límite superior
lim_inf = 0   #límite inferior
x = unif(lim_inf, lim_sup, cant_num)  #generamos números aleatorios en el intervalo

suma = 0 # inicializamos la variable de la suma

for i in range(cant_num):
    suma = suma + f
    
resultado = (lim_sup - lim_inf)*suma/cant_num

plt.xlabel('x')
plt.ylabel('y')
plt.plot(c, f)
plt.hist(x, density=True)

print('El resultado de la integral es: ')
print(resultado)
```

```python
import numpy as np    #Gráfica de la aproximación 

def mc_pi_aprox(N=10000):
    plt.figure(figsize=(8,8))  # tamaño de la figura
    x, y = np.random.uniform(0, 1, size=(2, N))
    interior = (x**2 + y**2) <= 1
    pi = interior.sum() * 4 / N
    error = abs((pi - np.pi) / pi) * 100
    exterior = np.invert(interior)
    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')
    plt.plot(0, 0, label='$\hat \pi$ = {:4.4f}\nerror = {:4.4f}%'
             .format(pi,error), alpha=0)
    plt.axis('square')
    plt.legend(frameon=True, framealpha=0.9, fontsize=16)

mc_pi_aprox()
```
