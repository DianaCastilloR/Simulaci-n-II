# Estimar el valor de pi usando la función random

```python
import random #La librería random contiene la función "uniform" devuelve un número aleatorio en el rango [a, b].

def aprox_pi(n):
    circle_points = 0  # Puntos dentro del cículo
    square_points = 0  # Puntos dentro del cuadrado
    
    for i in range(n):
        rand_x = random.uniform(-1,1)   # Generamos los puntos de x con la función "uniform" en (-1,1)
        rand_y = random.uniform(-1,1)   # Generamos los puntos de y con la función "uniform" en (-1,1)
        
        origin_dist =(rand_x**2 + rand_y**2)**0.5    # Distancia de cada punto del origen
        
        if origin_dist <= 1:             # Comprobamos si el punto está dentro del círculo
            circle_points += 1
            
        square_points += 1               # Contador de puntos en el cuadrado
        
        pi = 4*circle_points / square_points
        
    print("pi =", pi) 
    
```

```python
aprox_pi(100000)
```
